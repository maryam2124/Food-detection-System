## **Dataset Class Consolidation into Three Main Categories**

- Original food dataset contains multiple fine-grained food categories  
- Similar food items were merged into 3 main classes: Curry_Sabzi, Rice_Based, Sweet_Dessert  
- Class mapping was performed using strict label-level transformation logic  
- YOLO format annotations were preserved and reorganized into a unified structure  
- Final dataset was split into train, validation, and test sets for YOLOv8 training  

````python
# =========================
# INPUT / OUTPUT
# =========================
DATASET_ROOT = "/kaggle/input/datasets/mrimkhan/cooked/cooked"
OUTPUT_DIR = "/kaggle/working/merged_dataset"

# =========================
# FINAL CLASS DEFINITIONS (NO EXTRA CLASS)
# =========================
MERGE_MAP = {
    "Curry_Sabzi": {
        "BhindiMasala", "Chole", "dal", "ShahiPaneer", "palak_paneer"
    },
    "Rice_Based": {
        "Biryani", "rice", "poha"
    },
    "Sweet_Dessert": {
        "Jalebi", "gulab_jamun"
    }
}

FINAL_CLASSES = list(MERGE_MAP.keys())

# =========================
# CREATE OUTPUT STRUCTURE
# =========================
for split in ["train", "valid", "test"]:
    os.makedirs(f"{OUTPUT_DIR}/{split}/images", exist_ok=True)
    os.makedirs(f"{OUTPUT_DIR}/{split}/labels", exist_ok=True)


# =========================
# LOAD YAML
# =========================
def load_yaml(path):
    with open(path, "r") as f:
        return yaml.safe_load(f)


# =========================
# FIND DATASET
# =========================
def find_dataset():
    for root, dirs, files in os.walk(DATASET_ROOT):
        if "data.yaml" in files:
            return root
    return None


# =========================
# NORMALIZE NAME
# =========================
def norm(x):
    return x.strip()


# =========================
# PROCESS DATASET
# =========================
def process(dataset_path):
    data = load_yaml(os.path.join(dataset_path, "data.yaml"))

    old_names = data["names"]
    idx_to_name = {i: norm(n) for i, n in enumerate(old_names)}

    written = 0
    skipped = 0

    for split in ["train", "valid", "test"]:
        img_dir = os.path.join(dataset_path, split, "images")
        lbl_dir = os.path.join(dataset_path, split, "labels")

        if not os.path.exists(lbl_dir):
            continue

        for file in os.listdir(lbl_dir):
            if not file.endswith(".txt"):
                continue

            label_path = os.path.join(lbl_dir, file)
            new_labels = []
            valid = False

            with open(label_path, "r") as f:
                for line in f:
                    parts = line.strip().split()

                    if len(parts) != 5:
                        continue

                    try:
                        old_id = int(parts[0])
                    except:
                        continue

                    old_class = idx_to_name.get(old_id, "")

                    mapped_class = None

                    # STRICT MATCH ONLY
                    for new_cls, old_set in MERGE_MAP.items():
                        if old_class in old_set:
                            mapped_class = new_cls
                            break

                    if mapped_class is None:
                        continue

                    new_id = FINAL_CLASSES.index(mapped_class)

                    new_labels.append(
                        str(new_id) + " " + " ".join(parts[1:])
                    )
                    valid = True

            if valid:
                base = file.replace(".txt", "")

                # copy image
                for ext in [".jpg", ".png", ".jpeg"]:
                    img_path = os.path.join(img_dir, base + ext)
                    if os.path.exists(img_path):
                        shutil.copy(img_path, f"{OUTPUT_DIR}/{split}/images/")
                        break

                # write label
                with open(f"{OUTPUT_DIR}/{split}/labels/{file}", "w") as f:
                    f.write("\n".join(new_labels))

                written += 1
            else:
                skipped += 1

    print("\n========== PROCESS SUMMARY ==========")
    print("Written:", written)
    print("Skipped:", skipped)


# =========================
# RUN
# =========================
dataset_path = find_dataset()

if dataset_path is None:
    print("DATASET NOT FOUND")
else:
    process(dataset_path)


# =========================
# WRITE FINAL YAML
# =========================
yaml_out = {
    "train": "train/images",
    "val": "valid/images",
    "test": "test/images",
    "nc": len(FINAL_CLASSES),
    "names": FINAL_CLASSES
}

with open(f"{OUTPUT_DIR}/data.yaml", "w") as f:
    yaml.dump(yaml_out, f, sort_keys=False)

print("\nDONE")
print("OUTPUT:", OUTPUT_DIR)
````

````python
DATASET_PATH = "/kaggle/working/merged_dataset"
FINAL_CLASSES = ["Curry_Sabzi", "Rice_Based", "Sweet_Dessert"]

class_samples = defaultdict(list)

# =========================
# COLLECT 5 IMAGES PER CLASS
# =========================
for split in ["train", "valid", "test"]:
    lbl_dir = os.path.join(DATASET_PATH, split, "labels")
    img_dir = os.path.join(DATASET_PATH, split, "images")

    if not os.path.exists(lbl_dir):
        continue

    files = os.listdir(lbl_dir)
    random.shuffle(files)   # makes it random

    for file in files:
        label_path = os.path.join(lbl_dir, file)

        with open(label_path, "r") as f:
            lines = f.readlines()

        for line in lines:
            parts = line.strip().split()
            if len(parts) < 5:
                continue

            class_id = int(parts[0])

            if len(class_samples[class_id]) < 5:   # CHANGED
                base = file.replace(".txt", "")

                for ext in [".jpg", ".png", ".jpeg"]:
                    img_path = os.path.join(img_dir, base + ext)
                    if os.path.exists(img_path):
                        class_samples[class_id].append((img_path, label_path))
                        break

# =========================
# DISPLAY WITH GREEN BOXES
# =========================
for class_id, samples in class_samples.items():
    print(f"\nClass {class_id} ({FINAL_CLASSES[class_id]})")

    plt.figure(figsize=(15, 3))  # wider for 5 images

    for i, (img_path, label_path) in enumerate(samples):
        img = Image.open(img_path)
        w, h = img.size

        plt.subplot(1, 5, i+1)   #  5 images
        plt.imshow(img)
        ax = plt.gca()

        # Read labels
        with open(label_path, "r") as f:
            lines = f.readlines()

        for line in lines:
            parts = line.strip().split()
            if len(parts) < 5:
                continue

            cls = int(parts[0])
            x_center, y_center, bw, bh = map(float, parts[1:])

            # YOLO → pixel
            x1 = (x_center - bw/2) * w
            y1 = (y_center - bh/2) * h
            box_w = bw * w
            box_h = bh * h

            rect = plt.Rectangle(
                (x1, y1),
                box_w,
                box_h,
                fill=False,
                edgecolor="green",   #  GREEN BOX
                linewidth=2
            )

            ax.add_patch(rect)
            ax.text(
                x1, y1,
                FINAL_CLASSES[cls],
                color="green",       # GREEN TEXT
                fontsize=8,
                bbox=dict(facecolor='black', alpha=0.5)
            )

        plt.title(FINAL_CLASSES[class_id])
        plt.axis("off")

    plt.show()
````

## **Dataset Construction & Merging Strategy**

- Multiple independent YOLO-format food datasets were collected from different sources  
- Each dataset contained its own annotation schema and class naming convention  
- A unified global class list was defined to standardize all categories  
- Dataset labels were mapped from local class indices to global class indices  
- Inconsistent or ambiguous class names were normalized using a rule-based correction function  
- Invalid or corrupted YOLO annotations were filtered out during preprocessing  
- Only bounding boxes within valid normalized ranges were retained  
- All datasets were merged into a single structured dataset directory  
- Images and labels were reorganized into a consistent train/validation/test structure  
- A final consolidated `data.yaml` file was generated for YOLOv8 training  

````python
# =========================
# INPUT DATASETS
# =========================
DATASETS = [
    "/kaggle/input/datasets/mrimkhan/junkfood2/junkfood/hamburgers.v2i.yolov8",
    "/kaggle/input/datasets/mrimkhan/junkfood2/junkfood/pizza.v1i.yolov8",
    "/kaggle/input/datasets/mrimkhan/junkfood2/junkfood/j_donuts.v1i.yolov8",
    "/kaggle/input/datasets/mrimkhan/junkfood2/junkfood/Cooked Pasta Detection Yolov8.v5i.yolov8",
    "/kaggle/input/datasets/mrimkhan/junkfood2/junkfood/hot dog.v1i.yolov8",
    "/kaggle/input/datasets/mrimkhan/junkfood2/junkfood/Patates Kizartmasi.v1i.yolov8"
]

OUTPUT_DIR = "/kaggle/working/junk_food_updated"
SPLITS = ["train", "valid", "test"]

# =========================
# FINAL CLASSES
# =========================
GLOBAL_CLASSES = [
    "beef hamburger",
    "chicken-burger",
    "pizza",
    "donuts",
    "cooked_pasta",
    "hot dog",
    "fries"
]

class_to_id = {c: i for i, c in enumerate(GLOBAL_CLASSES)}

# =========================
# ONLY FIX NEEDED
# =========================
def fix_class(name):
    name = name.strip().lower()
    if name == "0":
        return "fries"
    return name

# =========================
# VALID YOLO CHECK
# =========================
def is_valid(parts):
    if len(parts) != 5:
        return False
    if not parts[0].isdigit():
        return False
    try:
        vals = list(map(float, parts[1:]))
        return all(0 <= v <= 1 for v in vals)
    except:
        return False

# =========================
# YAML LOADER
# =========================
def load_yaml(path):
    with open(path, "r") as f:
        data = yaml.safe_load(f)

    names = data.get("names", [])
    if isinstance(names, dict):
        names = [names[i] for i in sorted(names.keys())]

    return names

# =========================
# FIND YAML
# =========================
def find_yaml(path):
    for f in os.listdir(path):
        if f.endswith(".yaml") or f.endswith(".yml"):
            return os.path.join(path, f)
    return None

# =========================
# CREATE OUTPUT DIRS
# =========================
def create_dirs():
    for s in SPLITS:
        os.makedirs(os.path.join(OUTPUT_DIR, s, "images"), exist_ok=True)
        os.makedirs(os.path.join(OUTPUT_DIR, s, "labels"), exist_ok=True)

# =========================
# MERGE FUNCTION
# =========================
def merge():
    create_dirs()

    stats = {
        "written": 0,
        "skipped": 0
    }

    class_count = defaultdict(int)

    for dataset in DATASETS:
        print(f"\nProcessing: {dataset}")

        yaml_path = find_yaml(dataset)
        if not yaml_path:
            continue

        local_classes = load_yaml(yaml_path)

        for split in SPLITS:
            img_dir = os.path.join(dataset, split, "images")
            lbl_dir = os.path.join(dataset, split, "labels")

            if not os.path.exists(img_dir):
                continue

            for img in os.listdir(img_dir):

                img_path = os.path.join(img_dir, img)
                label_file = img.rsplit(".", 1)[0] + ".txt"
                label_path = os.path.join(lbl_dir, label_file)

                new_labels = []

                if os.path.exists(label_path):
                    with open(label_path, "r") as f:
                        for line in f:
                            parts = line.strip().split()

                            if not is_valid(parts):
                                continue

                            old_id = int(parts[0])

                            if old_id >= len(local_classes):
                                continue

                            class_name = local_classes[old_id]

                            # ONLY FIX
                            class_name = fix_class(class_name)

                            if class_name not in class_to_id:
                                continue

                            new_id = class_to_id[class_name]

                            new_labels.append(
                                " ".join([str(new_id)] + parts[1:])
                            )

                            class_count[class_name] += 1

                if len(new_labels) == 0:
                    stats["skipped"] += 1
                    continue

                out_img = os.path.join(OUTPUT_DIR, split, "images", img)
                out_lbl = os.path.join(OUTPUT_DIR, split, "labels", label_file)

                shutil.copy(img_path, out_img)

                with open(out_lbl, "w") as f:
                    f.write("\n".join(new_labels))

                stats["written"] += 1

    # =========================
    # FINAL YAML
    # =========================
    with open(os.path.join(OUTPUT_DIR, "data.yaml"), "w") as f:
        yaml.dump({
            "train": "train/images",
            "val": "valid/images",
            "test": "test/images",
            "nc": len(GLOBAL_CLASSES),
            "names": GLOBAL_CLASSES
        }, f, sort_keys=False)

    # =========================
    # REPORT
    # =========================
    print("\n================ FINAL REPORT ================")
    print("Written:", stats["written"])
    print("Skipped:", stats["skipped"])

    print("\nCLASS DISTRIBUTION:")
    for k, v in class_count.items():
        print(k, ":", v)


# =========================
# RUN
# =========================
merge()
````

````python

# =========================
# CONFIG
# =========================
DATASET_PATH = "/kaggle/working/junk_food_updated"
SPLITS = ["train", "valid", "test"]

# =========================
# LOAD YAML
# =========================
with open(os.path.join(DATASET_PATH, "data.yaml"), "r") as f:
    data = yaml.safe_load(f)

class_names = data["names"]

# =========================
# DRAW FUNCTION
# =========================
def draw_boxes(image_path, label_path):
    img = cv2.imread(image_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    h, w, _ = img.shape

    if os.path.exists(label_path):
        with open(label_path, "r") as f:
            lines = f.readlines()

        for line in lines:
            parts = line.strip().split()
            if len(parts) != 5:
                continue

            cls_id = int(parts[0])
            x, y, bw, bh = map(float, parts[1:])

            x1 = int((x - bw / 2) * w)
            y1 = int((y - bh / 2) * h)
            x2 = int((x + bw / 2) * w)
            y2 = int((y + bh / 2) * h)

            color = (0, 255, 0)

            cv2.rectangle(img, (x1, y1), (x2, y2), color, 2)

            label = class_names[cls_id] if cls_id < len(class_names) else str(cls_id)

            cv2.putText(img, label, (x1, y1 - 5),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)

    return img

# =========================
# COLLECT CLASS-WISE SAMPLES
# =========================
class_samples = defaultdict(list)

for split in SPLITS:
    img_dir = os.path.join(DATASET_PATH, split, "images")
    lbl_dir = os.path.join(DATASET_PATH, split, "labels")

    if not os.path.exists(img_dir):
        continue

    files = os.listdir(img_dir)
    random.shuffle(files)  #  RANDOM

    for img_name in files:
        if not img_name.lower().endswith((".jpg", ".jpeg", ".png")):
            continue

        img_path = os.path.join(img_dir, img_name)
        label_path = os.path.join(lbl_dir, img_name.rsplit(".", 1)[0] + ".txt")

        if not os.path.exists(label_path):
            continue

        with open(label_path, "r") as f:
            lines = f.readlines()

        for line in lines:
            parts = line.strip().split()
            if len(parts) != 5:
                continue

            cls_id = int(parts[0])

            if len(class_samples[cls_id]) < 4:  # ✅ 4 per class
                class_samples[cls_id].append((img_path, label_path))

# =========================
# DISPLAY (ROW = CLASS)
# =========================
num_classes = len(class_names)

plt.figure(figsize=(16, 4 * num_classes))

for class_id in range(num_classes):
    samples = class_samples[class_id]

    for i in range(4):
        plt.subplot(num_classes, 4, class_id * 4 + i + 1)

        if i < len(samples):
            img = draw_boxes(*samples[i])
            plt.imshow(img)
            plt.title(class_names[class_id])
        else:
            plt.text(0.5, 0.5, "No Image", ha='center')

        plt.axis("off")

plt.tight_layout()
plt.show()
````

## **Dataset Construction & Standardization Process**

- Multiple vegetable detection datasets were collected from different sources  
- Each dataset contained inconsistent class naming and annotation formats  
- A unified class schema was defined for six core vegetable categories  
- Class labels were normalized using rule-based mapping and synonym handling  
- Variations in spelling and naming (e.g., broccoli/brocoli) were corrected  
- Rare or inconsistent labels were mapped to standardized global classes  
- Invalid YOLO annotations were filtered during preprocessing  
- Each image-label pair was validated before inclusion in the final dataset  
- Images were renamed to avoid filename collisions across datasets  
- Final dataset was structured into train/validation/test splits  
- A unified `data.yaml` file was generated for YOLOv8 training compatibility  

````python
# =========================
# CONFIG
# =========================
DATASETS = [
    "/kaggle/input/datasets/mrimkhan/vegetables/vegetables/onion.v1i.yolov8",
    "/kaggle/input/datasets/mrimkhan/vegetables/vegetables/Sprouted potatoes.v2i.yolov8",
    "/kaggle/input/datasets/mrimkhan/vegetables/vegetables/cucumber.v1i.yolov8",
    "/kaggle/input/datasets/mrimkhan/vegetables/vegetables/brocoli.v3i.yolov8",
    "/kaggle/input/datasets/mrimkhan/vegetables/vegetables/carrot.v2i.yolov8",
    "/kaggle/input/datasets/mrimkhan/vegetables/vegetables/Tomato.v1i.yolov8"
]

OUTPUT_DIR = "/kaggle/working/vegetable_merged"
SPLITS = ["train", "valid", "test"]

# =========================
# FINAL CLASSES
# =========================
FINAL_CLASSES = [
    "onion",
    "potato",
    "cucumber",
    "broccoli",
    "carrot",
    "tomato"
]

class_to_id = {c: i for i, c in enumerate(FINAL_CLASSES)}

# =========================
# VALID YOLO LABEL
# =========================
def is_valid(parts):
    if len(parts) != 5:
        return False
    if not parts[0].isdigit():
        return False
    try:
        vals = list(map(float, parts[1:]))
        return all(0 <= v <= 1 for v in vals)
    except:
        return False

# =========================
# LOAD YAML
# =========================
def load_yaml(path):
    with open(path, "r") as f:
        data = yaml.safe_load(f)

    names = data.get("names", [])
    if isinstance(names, dict):
        names = [names[i] for i in sorted(names.keys())]

    return [n.strip().lower() for n in names]

# =========================
# FIND YAML
# =========================
def find_yaml(path):
    for f in os.listdir(path):
        if f.endswith(".yaml") or f.endswith(".yml"):
            return os.path.join(path, f)
    return None

# =========================
# CREATE DIRS
# =========================
def create_dirs():
    for s in SPLITS:
        os.makedirs(os.path.join(OUTPUT_DIR, s, "images"), exist_ok=True)
        os.makedirs(os.path.join(OUTPUT_DIR, s, "labels"), exist_ok=True)

# =========================
# CLASS NORMALIZATION
# =========================
def normalize_class(name):
    name = name.lower().strip()

    if name in ["sprouted potato", "potato", "potatoes"]:
        return "potato"

    if name in ["brocoli", "broccoli"]:
        return "broccoli"

    if name in ["wortel", "carrot"]:
        return "carrot"

    if name in ["onion"]:
        return "onion"

    if name in ["cucumber"]:
        return "cucumber"

    #  IMPORTANT: ALL tomato classes → tomato
    if name in ["0", "1", "2", "tomato"]:
        return "tomato"

    return None

# =========================
# MERGE
# =========================
def merge():
    create_dirs()

    stats = defaultdict(int)
    class_counts = defaultdict(int)

    for dataset in DATASETS:
        print(f"\n processing: {dataset}")

        yaml_path = find_yaml(dataset)
        if not yaml_path:
            print("No YAML, skipping")
            continue

        local_classes = load_yaml(yaml_path)

        for split in SPLITS:
            img_dir = os.path.join(dataset, split, "images")
            lbl_dir = os.path.join(dataset, split, "labels")

            if not os.path.exists(img_dir):
                continue

            for img_name in os.listdir(img_dir):

                img_path = os.path.join(img_dir, img_name)
                label_path = os.path.join(lbl_dir, img_name.rsplit(".",1)[0] + ".txt")

                new_labels = []

                if os.path.exists(label_path):
                    with open(label_path, "r") as f:
                        lines = f.readlines()

                    for line in lines:
                        parts = line.strip().split()

                        if not is_valid(parts):
                            stats["invalid"] += 1
                            continue

                        old_id = int(parts[0])

                        if old_id >= len(local_classes):
                            continue

                        raw_class = local_classes[old_id]
                        final_class = normalize_class(raw_class)

                        if final_class is None:
                            continue

                        new_id = class_to_id[final_class]

                        new_labels.append(
                            " ".join([str(new_id)] + parts[1:])
                        )

                        class_counts[final_class] += 1

                if len(new_labels) == 0:
                    stats["skipped"] += 1
                    continue

                unique_name = f"{dataset.split('/')[-1]}_{img_name}"

                # copy image
                shutil.copy(
                    img_path,
                    os.path.join(OUTPUT_DIR, split, "images", unique_name)
                )

                # write label
                with open(
                    os.path.join(OUTPUT_DIR, split, "labels",
                                 unique_name.rsplit(".",1)[0] + ".txt"),
                    "w"
                ) as f:
                    f.write("\n".join(new_labels))

                stats["written"] += 1

    # =========================
    # FINAL YAML
    # =========================
    with open(os.path.join(OUTPUT_DIR, "data.yaml"), "w") as f:
        yaml.dump({
            "train": "train/images",
            "val": "valid/images",
            "test": "test/images",
            "nc": len(FINAL_CLASSES),
            "names": FINAL_CLASSES
        }, f, sort_keys=False)

    # =========================
    # REPORT
    # =========================
    print("\n\n========== FINAL REPORT ==========")
    print("Written:", stats["written"])
    print("Skipped:", stats["skipped"])
    print("Invalid:", stats["invalid"])

    print("\n CLASS DISTRIBUTION:")
    for k, v in class_counts.items():
        print(f"{k}: {v}")

    print("\n OUTPUT:", OUTPUT_DIR)

# =========================
# RUN
# =========================
merge()
````

````python

# =========================
#  MAIN OUTPUT PATH (CHANGE THIS ONLY IF NEEDED)
# =========================
BASE_OUTPUT_DIR = "/kaggle/working/MEGA_FINAL_DATASET_MAIN"

SPLITS = ["train", "valid", "test"]

# =========================
# DATASETS (ADD ALL HERE)
# =========================
DATASETS = [
    "/kaggle/working/junk_food_updated",
    "/kaggle/working/fruits_merged",
    "/kaggle/working/vegetable_merged",
    "/kaggle/working/merged_dataset"
]

# =========================
# FINAL 21 CLASSES
# =========================
FINAL_CLASSES = [
    "beef hamburger", "chicken-burger", "pizza", "donuts",
    "cooked_pasta", "hot dog", "fries",

    "banana", "orange", "pomegranate", "mango", "grapes", "apple",

    "onion", "potato", "cucumber", "broccoli", "carrot", "tomato",

    "curry_sabzi", "rice_based", "sweet_dessert"
]

class_map = {c: i for i, c in enumerate(FINAL_CLASSES)}

# =========================
# VALIDATION
# =========================
def is_valid(parts):
    if len(parts) != 5:
        return False
    if not parts[0].isdigit():
        return False
    try:
        vals = list(map(float, parts[1:]))
        return all(0 <= v <= 1 for v in vals)
    except:
        return False

# =========================
# NORMALIZER (CRITICAL FIX)
# =========================
def normalize(name):
    name = name.lower().strip()

    # FAST FOOD
    if "hamburger" in name or "beef" in name:
        return "beef hamburger"
    if "chicken" in name:
        return "chicken-burger"
    if "pizza" in name:
        return "pizza"
    if "donut" in name:
        return "donuts"
    if "pasta" in name:
        return "cooked_pasta"
    if "hot" in name and "dog" in name:
        return "hot dog"
    if "fries" in name:
        return "fries"

    # FRUITS
    if "banana" in name:
        return "banana"
    if "orange" in name:
        return "orange"
    if "pomegranate" in name:
        return "pomegranate"
    if "mango" in name:
        return "mango"
    if "grape" in name:
        return "grapes"
    if "apple" in name:
        return "apple"

    # VEGETABLES (FIXED HERE 🔥)
    if "onion" in name:
        return "onion"

    if "potato" in name or name == "0" or "patates" in name or "sprouted potato" in name:
        return "potato"

    if "cucumber" in name:
        return "cucumber"
    if "broccoli" in name:
        return "broccoli"
    if "carrot" in name or "wortel" in name:
        return "carrot"
    if "tomato" in name:
        return "tomato"

    # COOKED FOOD
    if "curry" in name or "sabzi" in name:
        return "curry_sabzi"
    if "rice" in name:
        return "rice_based"
    if "dessert" in name or "sweet" in name:
        return "sweet_dessert"

    return None
# =========================
# HELPERS
# =========================
def load_yaml(path):
    with open(path, "r") as f:
        data = yaml.safe_load(f)
    names = data.get("names", [])
    return [str(n).strip() for n in names]

def find_yaml(path):
    for f in os.listdir(path):
        if f.endswith(".yaml") or f.endswith(".yml"):
            return os.path.join(path, f)
    return None

def create_dirs():
    for s in SPLITS:
        os.makedirs(os.path.join(BASE_OUTPUT_DIR, s, "images"), exist_ok=True)
        os.makedirs(os.path.join(BASE_OUTPUT_DIR, s, "labels"), exist_ok=True)

# =========================
# MERGE ENGINE
# =========================
def merge_all():
    create_dirs()

    stats = defaultdict(int)
    class_counts = defaultdict(int)

    for ds in DATASETS:
        print(f"\nProcessing: {ds}")

        yaml_path = find_yaml(ds)
        if not yaml_path:
            print("Missing YAML")
            continue

        local_classes = load_yaml(yaml_path)

        for split in SPLITS:
            img_dir = os.path.join(ds, split, "images")
            lbl_dir = os.path.join(ds, split, "labels")

            if not os.path.exists(img_dir):
                continue

            for img in os.listdir(img_dir):

                img_path = os.path.join(img_dir, img)
                label_path = os.path.join(lbl_dir, img.rsplit(".", 1)[0] + ".txt")

                new_labels = []

                if os.path.exists(label_path):
                    for line in open(label_path):
                        parts = line.strip().split()

                        if not is_valid(parts):
                            continue

                        old_id = int(parts[0])
                        if old_id >= len(local_classes):
                            continue

                        raw = local_classes[old_id]
                        cls = normalize(raw)

                        if cls is None:
                            stats["dropped"] += 1
                            continue

                        new_id = class_map[cls]
                        new_labels.append(" ".join([str(new_id)] + parts[1:]))

                        class_counts[cls] += 1

                if len(new_labels) == 0:
                    stats["skipped"] += 1
                    continue

                new_name = f"{ds.split('/')[-1]}_{img}"

                shutil.copy(
                    img_path,
                    os.path.join(BASE_OUTPUT_DIR, split, "images", new_name)
                )

                with open(
                    os.path.join(BASE_OUTPUT_DIR, split, "labels", new_name.rsplit('.',1)[0] + ".txt"),
                    "w"
                ) as f:
                    f.write("\n".join(new_labels))

                stats["written"] += 1

    # =========================
    # FINAL YAML
    # =========================
    with open(os.path.join(BASE_OUTPUT_DIR, "data.yaml"), "w") as f:
        yaml.dump({
            "train": "train/images",
            "val": "valid/images",
            "test": "test/images",
            "nc": len(FINAL_CLASSES),
            "names": FINAL_CLASSES
        }, f, sort_keys=False)

    # =========================
    # REPORT
    # =========================
    print("\n\n========== FINAL REPORT ==========")
    print("Written:", stats["written"])
    print("Skipped:", stats["skipped"])
    print("Dropped:", stats["dropped"])

    print("\nCLASS CHECK (21 CLASSES):")
    for c in FINAL_CLASSES:
        print(f"{c}: {class_counts[c]}")

# =========================
# RUN
# =========================
merge_all()
````

````python
DATASET = "/kaggle/working/MEGA_FINAL_DATASET_MAIN"
SPLITS = ["train", "valid", "test"]

# ---------------- TARGET BAD CLASSES ----------------
FIX_CLASSES = {
    3,   # donuts
    7,   # banana
    6,   # fries
    18,  # tomato
    17   # carrot
}

# ---------------- GET CLASS FROM LABEL ----------------
def get_class(label_path):
    try:
        with open(label_path, "r") as f:
            for line in f:
                parts = line.strip().split()
                return int(parts[0])
    except:
        return None

# ---------------- COLLECT ONLY BAD CLASS SAMPLES ----------------
def collect_bad_samples():
    data = []

    for split in SPLITS:
        lbl_dir = os.path.join(DATASET, split, "labels")
        img_dir = os.path.join(DATASET, split, "images")

        for file in os.listdir(lbl_dir):
            if not file.endswith(".txt"):
                continue

            label_path = os.path.join(lbl_dir, file)
            cls = get_class(label_path)

            if cls in FIX_CLASSES:
                base = file.replace(".txt", "")

                img = None
                for ext in [".jpg", ".jpeg", ".png"]:
                    p = os.path.join(img_dir, base + ext)
                    if os.path.exists(p):
                        img = p
                        break

                if img:
                    data.append((img, label_path, split, cls))

    return data

# ---------------- SAFE MOVE ----------------
def move_sample(img, lbl, src_split, dst_split):
    img_name = os.path.basename(img)
    lbl_name = os.path.basename(lbl)

    src_img = img
    src_lbl = lbl

    dst_img = os.path.join(DATASET, dst_split, "images", img_name)
    dst_lbl = os.path.join(DATASET, dst_split, "labels", lbl_name)

    if src_img == dst_img:
        return

    shutil.move(src_img, dst_img)
    shutil.move(src_lbl, dst_lbl)

# ---------------- BALANCE ----------------
def balance():
    samples = collect_bad_samples()

    print("Bad class samples found:", len(samples))

    random.shuffle(samples)

    # group per class
    per_class = {}

    for s in samples:
        per_class.setdefault(s[3], []).append(s)

    for cls, items in per_class.items():

        n = len(items)

        train_end = int(0.7 * n)
        val_end = int(0.9 * n)

        print(f"Class {cls} → {n} samples")

        for i, (img, lbl, src, _) in enumerate(items):

            if i < train_end:
                continue  # keep in train
            elif i < val_end:
                move_sample(img, lbl, src, "valid")
            else:
                move_sample(img, lbl, src, "test")

    print("CLASS FIX COMPLETED (ONLY BAD CLASSES)")

balance()
````

````python
# =========================
# MAKE RANDOM FIXED (IMPORTANT)
# =========================
random.seed(42)

DATASET_PATH = "/kaggle/working/MEGA_FINAL_DATASET_MAIN"
SPLITS = ["train", "valid", "test"]

LIMIT = 4000


# ---------------- COLLECT ----------------
def collect_data():
    class_counts = defaultdict(int)
    class_files = defaultdict(list)

    for split in SPLITS:
        label_dir = os.path.join(DATASET_PATH, split, "labels")

        if not os.path.exists(label_dir):
            continue

        for file in os.listdir(label_dir):
            if not file.endswith(".txt"):
                continue

            path = os.path.join(DATASET_PATH, split, "labels", file)

            with open(path, "r") as f:
                lines = f.readlines()

            file_classes = set()

            for line in lines:
                parts = line.strip().split()
                if len(parts) == 5 and parts[0].isdigit():
                    cls = int(parts[0])
                    file_classes.add(cls)
                    class_counts[cls] += 1

            for cls in file_classes:
                class_files[cls].append(path)

    return class_counts, class_files


# ---------------- DOWNSAMPLE ----------------
def downsample():
    class_counts, class_files = collect_data()

    remove_files = set()
    report = []

    print("\n================ DOWNSAMPLING REPORT ================\n")

    for cls, count in class_counts.items():

        if count > LIMIT:
            all_files = class_files[cls]

            keep_ratio = LIMIT / count
            keep_count = max(1, int(len(all_files) * keep_ratio))

            # FIXED RANDOMNESS (REPRODUCIBLE)
            random.seed(42)
            keep_files = set(random.sample(all_files, keep_count))

            remove = set(all_files) - keep_files
            remove_files.update(remove)

            report.append({
                "class": cls,
                "before": count,
                "after": count - sum(
                    1 for f in remove
                    for l in open(f)
                    if l.strip().split()[0].isdigit() and int(l.strip().split()[0]) == cls
                ),
                "removed_files": len(remove)
            })

            print(f"Class {cls} ↓ DOWNSAMPLED")
            print(f"   Before : {count}")
            print(f"   After  : ~{LIMIT}")
            print(f"   Files removed : {len(remove)}\n")

    # ---------------- DELETE ----------------
    for label_path in remove_files:
        img_path = label_path.replace("/labels/", "/images/").rsplit(".", 1)[0]

        if os.path.exists(label_path):
            os.remove(label_path)

        for ext in [".jpg", ".png", ".jpeg"]:
            if os.path.exists(img_path + ext):
                os.remove(img_path + ext)
                break

    # ---------------- SUMMARY ----------------
    print("\n================ FINAL SUMMARY ================\n")

    if not report:
        print("No class needed downsampling (all ≤ LIMIT)")
    else:
        for r in report:
            print(f"Class {r['class']}: removed {r['removed_files']} files")

    print("\nDone safely (DETERMINISTIC)")


# RUN
downsample()
````

````python
DATASET_PATH = "/kaggle/working/MEGA_FINAL_DATASET_MAIN"

SPLITS = ["train"]  # usually augment only train set

# ---------------- 4X TARGET CLASSES ----------------
AUG_CLASSES = {
    "carrot",
    "tomato",
    "fries",
    "chicken-burger",
    "mango"
}

# ---------------- AUGMENTATION PIPELINE ----------------
transform = A.Compose([
    A.HorizontalFlip(p=0.5),
    A.RandomBrightnessContrast(p=0.4),
    A.HueSaturationValue(p=0.4),
    A.ShiftScaleRotate(shift_limit=0.05, scale_limit=0.2, rotate_limit=20, p=0.5),
    A.Blur(blur_limit=3, p=0.2)
], bbox_params=A.BboxParams(format='yolo', label_fields=['class_labels']))

# ---------------- LOAD CLASS NAMES ----------------
import yaml
with open(os.path.join(DATASET_PATH, "data.yaml"), "r") as f:
    data = yaml.safe_load(f)

names = data["names"]
name_to_id = {n: i for i, n in enumerate(names)}

target_ids = {name_to_id[c] for c in AUG_CLASSES if c in name_to_id}


# ---------------- AUGMENT FUNCTION ----------------
def augment_image(img_path, label_path, save_img_dir, save_lbl_dir, count=4):

    image = cv2.imread(img_path)
    if image is None:
        return

    boxes = []
    class_labels = []

    if not os.path.exists(label_path):
        return

    with open(label_path, "r") as f:
        for line in f:
            parts = line.strip().split()
            if len(parts) != 5:
                continue

            # FIX 1: safe float conversion
            cls = int(float(parts[0]))

            if cls not in target_ids:
                return  # skip if image contains non-target class

            x, y, bw, bh = map(float, parts[1:])
            boxes.append([x, y, bw, bh])
            class_labels.append(cls)

    base_name = os.path.splitext(os.path.basename(img_path))[0]

    for i in range(count):
        try:
            augmented = transform(
                image=image,
                bboxes=boxes,
                class_labels=class_labels
            )

            aug_img = augmented["image"]
            aug_boxes = augmented["bboxes"]
            aug_labels = augmented["class_labels"]

            new_img_name = f"{base_name}_aug_{i}.jpg"
            new_lbl_name = f"{base_name}_aug_{i}.txt"

            cv2.imwrite(os.path.join(save_img_dir, new_img_name), aug_img)

            with open(os.path.join(save_lbl_dir, new_lbl_name), "w") as f:
                for cls, box in zip(aug_labels, aug_boxes):
                    f.write(f"{cls} {box[0]} {box[1]} {box[2]} {box[3]}\n")

        except:
            continue


# ---------------- RUN AUGMENTATION ----------------
for split in SPLITS:

    img_dir = os.path.join(DATASET_PATH, split, "images")
    lbl_dir = os.path.join(DATASET_PATH, split, "labels")

    for img_file in os.listdir(img_dir):

        if not img_file.endswith((".jpg", ".png", ".jpeg")):
            continue

        img_path = os.path.join(img_dir, img_file)
        lbl_path = os.path.join(lbl_dir, img_file.rsplit(".", 1)[0] + ".txt")

        augment_image(
            img_path,
            lbl_path,
            img_dir,
            lbl_dir,
            count=4
        )

print("4X augmentation completed")
````

````python
DATASET_PATH = "/kaggle/working/MEGA_FINAL_DATASET_MAIN"
SPLITS = ["train"]

# ---------------- 2X TARGET CLASSES ----------------
AUG_CLASSES = {
    "banana",
    "orange",
    "cucumber",
    "broccoli",
    "pomegranate"
}

# ---------------- LOAD YAML ----------------
with open(os.path.join(DATASET_PATH, "data.yaml"), "r") as f:
    data = yaml.safe_load(f)

names = data["names"]
name_to_id = {n: i for i, n in enumerate(names)}
target_ids = {name_to_id[c] for c in AUG_CLASSES if c in name_to_id}

# ---------------- AUGMENTATION PIPELINE ----------------
transform = A.Compose([
    A.HorizontalFlip(p=0.5),
    A.RandomBrightnessContrast(p=0.4),
    A.HueSaturationValue(p=0.3),
    A.Affine(
        translate_percent=0.05,
        scale=(0.85, 1.15),
        rotate=(-15, 15),
        p=0.5
    )
], bbox_params=A.BboxParams(format='yolo', label_fields=['class_labels']))

# ---------------- AUGMENT FUNCTION ----------------
def augment(img_path, lbl_path, img_dir, lbl_dir, count=2):

    if not os.path.exists(lbl_path):
        return

    image = cv2.imread(img_path)
    if image is None:
        return

    boxes = []
    labels = []

    with open(lbl_path, "r") as f:
        for line in f:
            parts = line.strip().split()

            if len(parts) != 5:
                continue

            # FIX: safe label parsing
            cls = int(float(parts[0]))

            # ONLY 2X TARGET CLASSES
            if cls not in target_ids:
                return

            x, y, w, h = map(float, parts[1:])
            boxes.append([x, y, w, h])
            labels.append(cls)

    base = os.path.splitext(os.path.basename(img_path))[0]

    for i in range(count):

        try:
            aug = transform(
                image=image,
                bboxes=boxes,
                class_labels=labels
            )

            aug_img = aug["image"]
            aug_boxes = aug["bboxes"]
            aug_labels = aug["class_labels"]

            if len(aug_boxes) == 0:
                continue

            new_img_path = os.path.join(img_dir, f"{base}_2x_{i}.jpg")
            new_lbl_path = os.path.join(lbl_dir, f"{base}_2x_{i}.txt")

            cv2.imwrite(new_img_path, aug_img)

            with open(new_lbl_path, "w") as f:
                for c, b in zip(aug_labels, aug_boxes):
                    f.write(f"{c} {b[0]} {b[1]} {b[2]} {b[3]}\n")

        except:
            continue


# ---------------- RUN ----------------
for split in SPLITS:
    img_dir = os.path.join(DATASET_PATH, split, "images")
    lbl_dir = os.path.join(DATASET_PATH, split, "labels")

    for img_file in os.listdir(img_dir):
        if not img_file.endswith((".jpg", ".png", ".jpeg")):
            continue

        img_path = os.path.join(img_dir, img_file)
        lbl_path = os.path.join(lbl_dir, img_file.rsplit(".", 1)[0] + ".txt")

        augment(img_path, lbl_path, img_dir, lbl_dir, count=2)

print("2X augmentation completed successfully")
````
